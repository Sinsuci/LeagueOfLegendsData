# Acquisition pseudo top 100

```python
import requests
from bs4 import BeautifulSoup
import fct
import re


url = input("Top 100 URL : ")
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.71 Safari/537.36', 'referer': 'https://...'}
result = requests.get(url, headers=headers)
page = result.content.decode()
soup = BeautifulSoup(page)
names_html = soup.find_all("span", attrs={"class":"name"})
names_list = []
for a in names_html :
    names_list.append(a.string)

all_player_list = []

for a in names_list :
    url = "https://www.leagueofgraphs.com/fr/summoner/euw/%s#championsData-soloqueue" %a
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.71 Safari/537.36', 'referer': 'https://...'}
    result = requests.get(url, headers=headers)
    page = result.content.decode()

    from bs4 import BeautifulSoup
    import fct

    soup = BeautifulSoup(page)
    soloq = soup.find_all("div", attrs={"data-tab-id":"championsData-soloqueue"})

    player_pseudo = [a]
    
    player_role_1 = []
    player_role_2 = []
    player_role_3 = []
    player_role_4 = []
    player_role_5 = []
    
    player_champ_1 = []
    player_champ_2 = []
    player_champ_3 = []
    player_champ_4 = []
    player_champ_5 = []

    player_list = [] #joueur : pseudo + overall + roles
    player_overall = [] #overall
    player_roles_all = [] #roles
    player_champ_bilan = [] #champs data

    player_list.append(player_pseudo)

    #Get data overall
    player_overall.append(fct.clean_txt(soloq[0].find_all("div", attrs={"id":re.compile("graphDD")})[0].string))
    player_overall.append(fct.clean_txt(soloq[0].find_all("div", attrs={"id":re.compile("graphDD")})[1].string))
    player_overall.append(fct.clean_txt(soloq[2].find("span", attrs={"class":"kills"}).string))
    player_overall.append(fct.clean_txt(soloq[2].find("span", attrs={"class":"deaths"}).string))
    player_overall.append(fct.clean_txt(soloq[2].find("span", attrs={"class":"assists"}).string))

    player_list.append(player_overall)

    #Get data role
    roles_data = soloq[3].find_all("td", attrs={"data-sort-value":True})
    if len(roles_data)/2 >= 1 :
        player_role_1.append(roles_data[0].previous_sibling.previous_sibling.find("img", attrs={"alt":True})['alt'])
        player_role_1.append(roles_data[0]['data-sort-value'])
        player_role_1.append(fct.winrateTO100(roles_data[0].next_sibling.next_sibling['data-sort-value']))
    if len(roles_data)/2 >= 2 :
        player_role_2.append(roles_data[2].previous_sibling.previous_sibling.find("img", attrs={"alt":True})['alt'])
        player_role_2.append(roles_data[2]['data-sort-value'])
        player_role_2.append(fct.winrateTO100(roles_data[2].next_sibling.next_sibling['data-sort-value']))
    if len(roles_data)/2 >= 3 :
        player_role_3.append(roles_data[4].previous_sibling.previous_sibling.find("img", attrs={"alt":True})['alt'])
        player_role_3.append(roles_data[4]['data-sort-value'])
        player_role_3.append(fct.winrateTO100(roles_data[4].next_sibling.next_sibling['data-sort-value']))
    if len(roles_data)/2 >= 4 :
        player_role_4.append(roles_data[6].previous_sibling.previous_sibling.find("img", attrs={"alt":True})['alt'])
        player_role_4.append(roles_data[6]['data-sort-value'])
        player_role_4.append(fct.winrateTO100(roles_data[6].next_sibling.next_sibling['data-sort-value']))
    if len(roles_data)/2 >= 5 :
        player_role_5.append(roles_data[8].previous_sibling.previous_sibling.find("img", attrs={"alt":True})['alt'])
        player_role_5.append(roles_data[8]['data-sort-value'])
        player_role_5.append(fct.winrateTO100(roles_data[8].next_sibling.next_sibling['data-sort-value']))
    player_roles_all.append(player_role_1)
    player_roles_all.append(player_role_2)
    player_roles_all.append(player_role_3)
    player_roles_all.append(player_role_4)
    player_roles_all.append(player_role_5)

    player_list.append(player_roles_all)

    #Get data champs
    player_champ_names = soloq[1].find_all("div", attrs={"class":"name"})
    player_champ_game_wr = soloq[1].find_all("progressbar", attrs={"data-value":True})
    if len(player_champ_names) >= 1 :
        player_champ_1.append(player_champ_names[0].string) #Nom
        player_champ_1.append(player_champ_game_wr[0]['data-value']) #Game
        player_champ_1.append(fct.winrateTO100(player_champ_game_wr[1]['data-value'])) #Winrate
    if len(player_champ_names) >= 2 :
        player_champ_2.append(player_champ_names[1].string)
        player_champ_2.append(player_champ_game_wr[2]['data-value'])
        player_champ_2.append(fct.winrateTO100(player_champ_game_wr[3]['data-value']))
    if len(player_champ_names) >= 3 :
        player_champ_3.append(player_champ_names[2].string)
        player_champ_3.append(player_champ_game_wr[4]['data-value'])
        player_champ_3.append(fct.winrateTO100(player_champ_game_wr[5]['data-value'])) 
    if len(player_champ_names) >= 4 :
        player_champ_4.append(player_champ_names[3].string)
        player_champ_4.append(player_champ_game_wr[6]['data-value'])
        player_champ_4.append(fct.winrateTO100(player_champ_game_wr[7]['data-value']))
    if len(player_champ_names) >= 5 :
        player_champ_5.append(player_champ_names[4].string)
        player_champ_5.append(player_champ_game_wr[8]['data-value'])
        player_champ_5.append(fct.winrateTO100(player_champ_game_wr[9]['data-value']))
    player_champ_bilan.append(player_champ_1)
    player_champ_bilan.append(player_champ_2)
    player_champ_bilan.append(player_champ_3)
    player_champ_bilan.append(player_champ_4)
    player_champ_bilan.append(player_champ_5)

    player_list.append(player_champ_bilan)

    all_player_list.append(player_list)


print("Finished")
```

```python
#Main role / OTP / Error
main_list = []

trigger_main = 60
trigger_otp = 90

main_role_counter = 0
otp_role_counter = 0
nomain_role_counter = 0

for a in all_player_list :
    if "." in a[1][0] :
        print("erreur")
    else :
        if fct.winrateTO100(int(a[2][0][1]) / int(a[1][0])) >= trigger_otp :
            print(a[0][0],"   otp in role :   ", a[2][0][0])
            otp_role_counter = otp_role_counter + 1
        elif fct.winrateTO100(int(a[2][0][1]) / int(a[1][0])) >= trigger_main :
            print(a[0][0],"   main in role :   ", a[2][0][0])
            main_role_counter = main_role_counter + 1
        else :
            print(a[0][0],"   no main")
            nomain_role_counter = nomain_role_counter + 1

print('\n',"main : ",main_role_counter,'\n',"otp : ",otp_role_counter,'\n',"Nothing : ",nomain_role_counter)


#graph main / otp / error nbr
import matplotlib.pyplot as plt

x_graph = ["Main","OTP","Nothing"]
y_graph = [main_role_counter,otp_role_counter,nomain_role_counter]

plt.bar(x_graph, y_graph)
plt.suptitle('Top 100 EUW role')
plt.savefig("Top 100 EUW RoleOTP.pdf", dpi = 600)
plt.show()
```

```python
#Top 100 EUW Game Played / Role
mid_role_counter = 0
jgl_role_counter = 0
adc_role_counter = 0
top_role_counter = 0
supp_role_counter = 0

trigger = 0.5

for a in all_player_list :
    if fct.winrateTO100(int(a[2][0][1]) / int(a[1][0])) >= trigger :
        for i in a[2] :
            if len(i) > 0 :
                if i[0] == "Mid" :
                    mid_role_counter = mid_role_counter + int(i[1])
                elif i[0] == "Support" :
                    supp_role_counter = supp_role_counter + int(i[1])
                elif i[0] == "Top" :
                    top_role_counter = top_role_counter + int(i[1])
                elif i[0] == "Jungler" :
                    jgl_role_counter = jgl_role_counter + int(i[1])
                elif i[0] == "ADC" :
                    adc_role_counter = adc_role_counter + int(i[1])


#horizontal graph main / otp roles
import matplotlib.pyplot as plt

x_graph = ["Game TOP","Game MID","Game JGL","Game ADC","Game SUPP"]
y_graph = [top_role_counter,mid_role_counter,jgl_role_counter,adc_role_counter,supp_role_counter]

plt.barh(x_graph, y_graph,)
plt.suptitle('Top 100 EUW GamePlayed per Role')
plt.savefig("Top 100 EUW GamePlayed per Role.pdf", dpi = 600)
plt.show()
```

```python
#Champion + compteur d'OTP sur un ROLE

champ_list = []
otp_counter = []

trigger_otp_game = 75
trigger_otp_champ = 75

lane = "Support"



for a in all_player_list :
    if fct.winrateTO100(int(a[2][0][1]) / int(a[1][0])) >= trigger_otp_game :
        if a[2][0][0] == "%s"%lane : #Vérifie bien un OTP ROLE
            if fct.winrateTO100(int(a[3][0][1]) / int(a[1][0])) >= trigger_otp_champ : #Vérifie bien un OTP champion
                line = -1
                for b in champ_list :
                    line = line + 1
                    if b == a[3][0][0] :
                        otp_counter[line] = int(otp_counter[line]) + int(1)
                
                if a[3][0][0] in champ_list :
                    a[0] #osef
                else :
                    champ_list.append(a[3][0][0])
                    otp_counter.append("1")

if len(otp_counter) or len(champ_list) != 0 :
    #Trie en fct des plus joués
    list1 = otp_counter
    list1 = list(map(int, list1))
    list2 = champ_list

    zipped_lists = zip(list1, list2)
    sorted_pairs = sorted(zipped_lists,reverse=True)

    tuples = zip(*sorted_pairs)
    list1, list2 = [ list(tuple) for tuple in  tuples]


    import numpy as np
    import pandas as pd
    import matplotlib.pyplot as plt

    font = {'family' : 'normal',
            'weight' : 'bold',
            'size'   : 4}
    plt.rc('font', **font)


    data = {'Champions': list2, 'Nbr of OTP': np.array(list1).astype(float) }

    df = pd.DataFrame(data,columns=['Champions','Nbr of OTP'])
    df.plot(x ='Champions', y='Nbr of OTP', kind = 'barh')

    specs1 = "OTP games on role : %s÷" %trigger_otp_game + "    OTP games on champ : %s÷" %trigger_otp_champ
    specs2 = "_%s_" %lane + "_role%s÷" %trigger_otp_game + "_champ%s÷" %trigger_otp_champ

    plt.suptitle('Top 100 EUW champ OTP played in %s'%lane)
    plt.title(specs1)

    plt.savefig("top 100 EUW champ OTP played in %s.pdf"%specs2, dpi = 600)
    plt.show()
else :
    print("empty")
```
